```python
def getNumFile(file):
    path = "\\".join(str(file).split('\\')[:-1])
    actual_filenames = ["_".join(str(i).split("_")[1:]) for i in os.listdir('./opus_files')]
    n = actual_filenames.count(file.name)
    print(file.name)
    print(n)
    new_filename = f"{n}_"+(file.name)
    return new_filename, str(path)
```
